```python
from aocd import get_data, submit
import numpy as np


data = get_data(year=2024, day=4).splitlines()
data
```

```python
xmases = 0

for line in data:
    if "XMAS" in line:
        xmases += line.count("XMAS")
    if "SAMX" in line:
        xmases += line.count("SAMX")

transposed = list(np.array([list(line) for line in data]).transpose())
transposed = ["".join(line) for line in transposed]

for line in transposed:
    if "XMAS" in line:
        xmases += line.count("XMAS")
    if "SAMX" in line:
        xmases += line.count("SAMX")

# Diagonals

for a in range(0, len(data) - 3):
    for i, j, k, l in zip(
        range(0, len(data)),
        range(1, len(data)),
        range(2, len(data)),
        range(3, len(data)),
    ):
        if "XMAS" in data[i][a] + data[j][a + 1] + data[k][a + 2] + data[l][a + 3]:
            xmases += 1
        if "SAMX" in data[i][a] + data[j][a + 1] + data[k][a + 2] + data[l][a + 3]:
            xmases += 1

for a in range(0, len(data) - 3):
    for i, j, k, l in zip(
        range(3, len(data)),
        range(2, len(data)),
        range(1, len(data)),
        range(0, len(data)),
    ):
        if "XMAS" in data[i][a] + data[j][a + 1] + data[k][a + 2] + data[l][a + 3]:
            xmases += 1
        if "SAMX" in data[i][a] + data[j][a + 1] + data[k][a + 2] + data[l][a + 3]:
            xmases += 1


print(xmases)
```

```python
submit(xmases, part="a", year=2024, day=4)
```

```python
mases = 0

# Part 2

for l1, l2, l3 in zip(range(0, len(data)), range(1, len(data)), range(2, len(data))):
    for a, b, c in zip(range(0, len(data)), range(1, len(data)), range(2, len(data))):
        if data[l2][b] == "A":
            if (data[l1][a] == "S" and data[l3][c] == "M") or (
                data[l1][a] == "M" and data[l3][c] == "S"
            ):
                if (data[l3][a] == "S" and data[l1][c] == "M") or (
                    data[l3][a] == "M" and data[l1][c] == "S"
                ):
                    mases += 1

mases
```

```python
submit(mases, part="b", year=2024, day=4)
```
